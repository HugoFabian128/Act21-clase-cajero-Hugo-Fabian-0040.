```python
class Cajero:
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    def marcar(self):
        print(f"{self.nombre} está marcando los productos.")

    def cobrar(self):
        print(f"{self.nombre} está cobrando la cuenta.")

    def pagar(self):
        print(f"{self.nombre} está entregando el cambio y el ticket.")

# Crear objetos
cajero1 = Cajero("Carlos Ruiz", "555-1234", 28)
cajero2 = Cajero("Ana López", "555-5678", 24)

# Ejecutar funciones Cajero 1
print(f"Cajero 1: {cajero1.nombre}, {cajero1.telefono}, {cajero1.edad} años")
cajero1.marcar()
cajero1.cobrar()
cajero1.pagar()

print("-" * 20)

# Ejecutar funciones Cajero 2
print(f"Cajero 2: {cajero2.nombre}, {cajero2.telefono}, {cajero2.edad} años")
cajero2.marcar()
cajero2.cobrar()
cajero2.pagar()
print("programa terminado")
print("Creador de software por Hugo Fabian NL = 15")
```
